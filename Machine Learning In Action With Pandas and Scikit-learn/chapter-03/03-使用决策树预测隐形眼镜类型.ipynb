```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
np.set_printoptions(precision=4)
```

```python
dataSet_df = pd.read_table('lenses.txt', names=['feature_1', 'feature_2', 'feature_3', 'feature_4', 'label'])
dataSet_df
```

```python
dummy_x = pd.get_dummies(dataSet_df.iloc[:, :-1])
dummy_y = pd.get_dummies(dataSet_df.iloc[:, -1])
dummy_x
```

```python
dummy_y
```

```python
# 生成 23 个随机索引
indices = np.random.permutation(dataSet_df.shape[0]) 

# 选取 20 个数据为 训练集
dummy_x_train = dummy_x.iloc[:, :].values[indices[:-3]]
dummy_y_train = dummy_y.iloc[:, :].values[indices[:-3]]

# 选取 3 个数据为 测试集
dummy_x_test = dummy_x.iloc[:, :].values[indices[-3:]]
dummy_y_test = dummy_y.iloc[:, :].values[indices[-3:]]

from sklearn.tree import DecisionTreeClassifier
# entropy 决定了用 ID3 算法 
dec_tree = DecisionTreeClassifier(criterion='entropy')
dec_tree.fit(dummy_x_train, dummy_y_train)
```

```python
dec_tree.predict(dummy_x_test)
```

```python
dec_tree.predict_proba(dummy_x_test)
```

```python
dec_tree.score(dummy_x_test, dummy_y_test)
```
